```python
##Remove / value in rate column
```

```python
def Handlerate(value):
    value = str(value).split("/")
    value = value[0]
    return float(value)

dataframe["rate"] = dataframe["rate"].apply(Handlerate)
```

```python
##what type of resturant do the majority of customers order from ?
```

```python
# Use 'listed_in(type)' for hue to avoid the warning
plt.figure(figsize=(7, 5))
sns.countplot(x='listed_in(type)', data=dataframe, palette='dark', hue='listed_in(type)')



# Add axis labels
plt.xlabel("Type of Restaurant")
plt.show()
```

```python
# How many votes has each type of resturant received from customer
```

```python
dataframe.head()
```

```python
df = dataframe.groupby("listed_in(type)") ["votes"].sum()
dfe = pd.DataFrame({'votes': df})
plt.plot(dfe , marker = 'o' , c= "green")
plt.xlabel("Type" , c = "red" , size = 20)
plt.ylabel ("Votes" , c = "red" , size = 20)
```
